```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.feature_selection import RFE
from sklearn.datasets import make_regression
```

```python
T = final_df.copy()
```

```python
T
```

```python
from sklearn.preprocessing import MinMaxScaler


# Initialize the scaler
scaler = MinMaxScaler()

# Fit and transform the data
scaled_data = scaler.fit_transform(T)

# Convert the scaled data back to a DataFrame
scaled_df = pd.DataFrame(scaled_data, columns=T.columns)
```

```python
scaled_df.columns
```

```python
# Separate the features and target variable
X = scaled_df.drop('crime rate per 100k Population', axis=1)
y = scaled_df['crime rate per 100k Population']

# Recursive Feature Elimination (RFE)
model = LinearRegression()

# Create an RFE object and specify the number of features to select
rfe = RFE(estimator=model, n_features_to_select=5)

# Fit RFE to the data
X_rfe = rfe.fit_transform(X, y)

# Get the selected features
selected_features_rfe = X.columns[rfe.support_]
print("Selected Features using RFE:")
print(selected_features_rfe)
```

```python
# LASSO Regularization
lasso = Lasso(alpha=0.1)

# Fit Lasso to the data
lasso.fit(X, y)

# Get the coefficients and select the features with non-zero coefficients
selected_features_lasso = X.columns[lasso.coef_ != 0]
print("Selected Features using LASSO:")
print(selected_features_lasso)
```

```python
from sklearn.decomposition import PCA
```

```python
# Perform PCA
pca = PCA()
pca.fit(scaled_df)

# Calculate the explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_

# Determine the number of components to retain based on explained variance
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)
num_components = np.argmax(cumulative_variance_ratio >= 0.95) + 1

# Apply PCA with the selected number of components
pca = PCA(n_components=num_components)
selected_features = pca.fit_transform(scaled_df)

# Print the selected features
print(selected_features)
```

```python
# Step 2: Print explained variance ratio
for i, ratio in enumerate(explained_variance_ratio):
    print(f"Principal Component {i+1}: {ratio:.3f}")

# Step 3: Analyze loadings
loadings = pca.components_
for i, loading in enumerate(loadings):
    print(f"Principal Component {i+1} Loadings:")
    for j, feature in enumerate(list(T.columns)):
        print(f"{feature}: {loading[j]:.3f}")
```
